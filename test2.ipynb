# 공격자산
- QQQ, VWO, VEA, BND
- 일별 수익률, 누적 수익률


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 


def to_dataframe(path, name):
    df = pd.read_csv(f'./data/{path}/daily/{name}.csv')[['Date','Price']].loc[::-1].set_index(['Date'], drop=True)
    return df
```

```python
bnd_df = to_dataframe('aggressive', 'BND')
qqq_df = to_dataframe('aggressive', 'QQQ')
vea_df = to_dataframe('aggressive', 'VEA')
vwo_df = to_dataframe('aggressive', 'VWO')

# 일별 수익률
bnd_df['Daily_profit'] = bnd_df['Price'].pct_change()
qqq_df['Daily_profit'] = qqq_df['Price'].pct_change()
vea_df['Daily_profit'] = vea_df['Price'].pct_change()
vwo_df['Daily_profit'] = vwo_df['Price'].pct_change()

# 누적 수익률
bnd_df['Profit_acc'] = (1+bnd_df['Daily_profit']).cumprod()-1
qqq_df['Profit_acc'] = (1+qqq_df['Daily_profit']).cumprod()-1
vea_df['Profit_acc'] = (1+vea_df['Daily_profit']).cumprod()-1
vwo_df['Profit_acc'] = (1+vwo_df['Daily_profit']).cumprod()-1

data = {'BND': bnd_df, 'QQQ': qqq_df, 'VEA': vea_df, 'VWO': vwo_df}

bnd_df['Profit_acc'].plot(label='BND')
qqq_df['Profit_acc'].plot(label='QQQ')
vea_df['Profit_acc'].plot(label='VEA')
vwo_df['Profit_acc'].plot(label='VWO')
plt.legend()
```

# 안전자산
