# 공격자산
- QQQ, VWO, VEA, BND
- 일별 수익률, 누적 수익률


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from backtesting import Backtest, Strategy


def to_dataframe(path, name):
    df = pd.read_csv(f'./data/{path}/daily/{name}.csv')[['Date','Price']].loc[::-1].set_index(['Date'], drop=True)
    return df


def profit_acc(path, name):
    df = to_dataframe(path, name)
    df['Daily_profit'] = df['Price'].pct_change()
    df['Profit_acc'] = (1 + df['Daily_profit']).cumprod() - 1 

    return df['Profit_acc']


class BAAStrategy(Strategy):
    pass
```
